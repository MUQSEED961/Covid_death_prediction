```python
from sklearn.preprocessing import StandardScaler
```

```python
stand=StandardScaler()
```

```python
x_covid = stand.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_covid,y,test_size=0.2,random_state=45)
y_train = y_train.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)
```

```python
import tensorflow as tf
```

```python
model1=tf.keras.models.Sequential()
```

```python
model1
```

```python
#layer1
model1.add(tf.keras.layers.Dense(units=10,activation='relu'))
```

```python
#layer2
model1.add(tf.keras.layers.Dense(units=10,activation='relu'))
```

```python
#output
model1.add(tf.keras.layers.Dense(units=1))
```

```python
#compile
model1.compile(optimizer='adam',loss='mean_squared_error',metrics=['mean_absolute_error'])
```

```python
model1.fit(x_train,y_train,epochs=1,batch_size=45)
```

```python
predictions = model1.predict(x_test)
y_pred1 = predictions
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import r2_score, root_mean_squared_error

print('R2 Score:', r2_score(y_test, y_pred1))
print('Root Mean Squared Error:', root_mean_squared_error(y_test, y_pred1))
```
